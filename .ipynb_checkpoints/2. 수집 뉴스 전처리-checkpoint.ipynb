```python
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pymysql
import os
import re
```

### 수집된 뉴스 파일 목록 확인

```python
arrs = os.listdir('D:\python\output\data')
arrs
```

### 수집된 뉴스 파일 목록

```python
pd.read_csv('D:\python\output\data\Article_경제_201601_201603.csv',encoding='cp949',names=['날짜','분야','언론사','제목','내용','링크'])
```

### 수집된 뉴스 낱개 파일을 하나로 병합

```python
df = pd.DataFrame()
for i in arrs: 
    df_s = pd.read_csv('D:\python\output\data\{}'.format(i),encoding='cp949',names=['날짜','분야','언론사','제목','내용','링크']) 
    df = df.append(df_s)
```

### 특정 기업과 관련된 뉴스만을 선별
### 제목에 '삼성전자' 라는 단어가 포함된 뉴스를 선별함

```python
df = df[df['제목'].apply(lambda x:'삼성전자' in x)] 
df
```

### 감성 분석에 뉴스의 제목만 사용할 예정

```python
df = df[['날짜','제목']] 
df
```

### 제목을 활용해 중복 뉴스를 제거하고 날짜를 datetime 형식으로 변환함

```python
df = df.drop_duplicates(subset=['제목']) # 제목을 활용해 중복 뉴스 제거
df['날짜'] = df['날짜'].apply(lambda x:(x[:11]+' '+x[-5:]+' '+x[-8:-5]).replace('.  ','. ').replace('오후','PM').replace('오전','AM').replace('  ',' '))
df['날짜'] = df['날짜'].apply(lambda x:pd.to_datetime(x))  
df
```

### 뉴스 제목에 포함된 특수문자 제거

```python
df['제목'] =  df['제목'].apply(lambda x: re.sub('[^A-Za-z0-9가-힣 ]', '', x).strip())
```

### 각 뉴스 헤드라인의 길이를 시각화함

```python
df['len'] = df['제목'].apply(lambda x:len(x)) 
plt.boxplot(df.len)
```

### 헤드라인 글자수 20 미만의 뉴스를 버리고, 길이가 긴 헤드라인은 50자까지만 사용함

```python
df = df[df['len']>=20] 
df['제목'] = df['제목'].apply(lambda x:x[:50])
```

```python
df
```

### 오후 3시 30분 이후로는 장마감 되기 때문에 이때의 뉴스는 다음날 주가에 영향을 미침  
### 따라서 해당 뉴스의 날짜를 다음날로 수정해주고, 시간 정보를 없엠

```python
df['날짜'] = df['날짜'].apply(lambda x: x+timedelta(1) if x.hour >= 15 and x.minute>=30 else x)
df['날짜'] = df['날짜'].apply(lambda x:pd.to_datetime(str(x)[:10]))
```

### 전처리 파일 저장

```python
df.to_excel('삼성전자 뉴스.xlsx')
```
