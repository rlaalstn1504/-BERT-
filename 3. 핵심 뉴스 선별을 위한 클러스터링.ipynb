```python
import os
import re
import datetime
import numpy as np
import time
from tqdm import tqdm  

import tensorflow as tf
from transformers import *

from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sentence_transformers import SentenceTransformer 

import pandas as pd
import matplotlib.pyplot as plt 

import umap # 차원 축소
import hdbscan # 클러스터링 
import pymysql
```

### 날짜별 전체  뉴스 파일 로드 

```python
df = pd.read_excel('삼성전자 뉴스.xlsx',index_col=0) 
df
```

#### - 각 뉴스는 주가의 변화량으로 라벨링이 될 예정임 

#### - 그러나 같은 날짜의 서로 다른 뉴스에 동일한 라벨을 부여하면 학습이 잘 안될 것으로 예상됨

#### - 따라서 날짜별 뉴스의 제목을 클러스터링 해서 가장 큰 군집을 형성한 뉴스 중 1개를 추출하여 학습에 사용함  

```python
# SentenceTransformer를 활용한 버트 임베딩 벡터를 이용해 클러스터링을 진행할 예정임
model = SentenceTransformer('distilbert-base-nli-mean-tokens') 

def get_cluster_label(df,column): 
    embeddings = model.encode(df[column].values)     
    umap_embeddings = umap.UMAP(n_neighbors=5, #작을수록 미세군집 형성 확률 커질?
                            n_components=5, #차원수 
                            min_dist=0.0,
                            metric='cosine').fit_transform(embeddings) 

    cluster = hdbscan.HDBSCAN(min_cluster_size=10,
                          min_samples=1, # 클러스터의 최소 크기
                          metric='euclidean',                      
                          cluster_selection_method='eom', 
                          prediction_data=True).fit(umap_embeddings)      
    return cluster.labels_
```

```python
new_df = pd.DataFrame() 
for i in tqdm(df.날짜.unique()): 
    try:
        one_df = df[df['날짜']==i] 
        one_df['label'] = get_cluster_label(one_df,'제목') # 임베딩벡터화 ㅡ> 차원축소 ㅡ> 클러스터링 레이블 출력
        largest_cluster = one_df['label'].value_counts().keys()[0] 
        new_df = new_df.append(one_df[one_df['label']==largest_cluster].sample(1)) 
    # 뉴스가 적은 날은 클러스터링이 안됨. 따라서 1개뉴스 랜덤 추출
    except: 
        one_df['label'] = -1
        new_df = new_df.append(one_df.sample(1))
```

```python
new_df.to_excel('클러스터링완료된삼성전자.xlsx')
```
